```python
def binarnaPretragaRekurzivno(lista, start, end, brojKojiTrazimo): 

    mid = (start + end) // 2 

    if lista[mid] == brojKojiTrazimo: 
        return mid 
    elif lista[mid] > brojKojiTrazimo: 
        return binarnaPretragaRekurzivno(lista, start, mid - 1, brojKojiTrazimo) 
    else: 
        return binarnaPretragaRekurzivno(lista, mid + 1, end, brojKojiTrazimo)
```

```python
lista = [2,5,7,9,11,13,15,17,22,25] 
brojKojiTrazimo = 13
r = binarnaPretragaRekurzivno(lista,0, len(lista) - 1,  brojKojiTrazimo) 
print("Broj koji trazimo se nalazi na indeksu " , r)
```
