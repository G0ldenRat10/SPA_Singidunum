```python
def selection_sort(nasa_lista): 
    n = len(nasa_lista) 

    for i in range(n): 
        min_index = i 
        for j in range(i + 1, n): 
            if nasa_lista[j] < nasa_lista[min_index]: 
                min_index = j 
        nasa_lista[i], nasa_lista[min_index] = nasa_lista[min_index], nasa_lista[i]
```

```python
nasa_lista = [55,11,44,22,66,33] 
print("Originalna lista", nasa_lista) 
selection_sort(nasa_lista) 
print("Sortirana lista", nasa_lista) 
```
