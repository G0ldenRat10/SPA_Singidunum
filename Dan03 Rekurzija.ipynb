```python
def faktorijel(x): 
    if x == 1: 
        return 1 
    else: 
        return(x * faktorijel(x - 1))
```

```python
faktorijel(10)
```

```python
def stepenovanje(broj, stepen): 
    if stepen == 0: 
        return 1 
    else: 
        return broj*stepenovanje(broj, stepen - 1)
```

```python
stepenovanje(12,3)
```

```python
# 5387 -> 23 

def sum_digits(n): 
    if n == 0: 
        return 0 
    return n % 10 + sum_digits(n // 10)
```

```python
print(sum_digits(5387))
```

```python
# [1,2,3] -> sve permutacije 
def permutations(lista): 
    if len(lista) == 0: 
        return [[]] 
    result = [] 

    for i in range(len(lista)): 
        elem = lista[i] 
        rest = lista[:i] + lista[i + 1:] 

        for p in permutations(rest): 
            result.append([elem] + p)
    return result
```

```python
print(permutations([1,2,3,4]))
```
