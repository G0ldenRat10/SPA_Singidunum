## Bubble Sort

Pseudo Kod

Izvrsni kod

```python
def bubble_sort(arr): 
    n = len(arr)
    for i in range (n): 
        has_swapped = False 
        for j in range(0, n - 1 - i): 
            if arr[j] > arr[j + 1]: 
                #Swap 
                arr[j], arr[j + 1] = arr[j + 1], arr[j] 
                has_swapped = True 
            if not has_swapped: 
                continue
```

```python
nasa_lista = [55,22,44,33,11,66] 
print("Originalna lista", nasa_lista) 
bubble_sort(nasa_lista) 
print("Sortirana lista", nasa_lista)
```
