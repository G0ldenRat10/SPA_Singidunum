```python
def karatsuba(x, y): 
    if len(str(x)) == 1 or len(str(y)) == 1: 
        return x * y 
    else: 
        x = str(x) 
        y = str(y) 
        
        n = max(len(x), len(y)) 
        half = n//2 
        
        a = int(x)/10**(half) 
        b = int(x) % 10**(half) 
        
        c = int(y)/10**(half) 
        d = int(y) % 10**(half) 
        
        a = int(a) 
        b = int(b) 
        c = int(c) 
        d = int(d) 
        
        p = a + b 
        q = c + d 
        
        ac = karatsuba(a, c) 
        bd = karatsuba(b, d) 
        pq = karatsuba(p, q) 
        
        adbc = pq - ac - bd 
        
        proizvod = ac*10**(2*half) + (adbc * 10**half) + bd 
        return proizvod
        
```

```python
karatsuba(5678, 1234)
```
