## Linearna Pretraga

```python
def linearnaPretraga(lista, vrednostKojuTrazimo): 
    for i in range(len(lista)): 
        if lista[i] == vrednostKojuTrazimo: 
            return i 
    return -1
```

```python
lista = [1,11,2,12,3,13,4,14,5,15] 
vrednostKojuTrazimo = 13 
linearnaPretraga(lista, vrednostKojuTrazimo)
```
