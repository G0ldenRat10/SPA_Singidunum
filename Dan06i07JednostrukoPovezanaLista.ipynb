```python
class Node: 
    def __init__(self, data, next_p = None): 
        self.data = data 
        self.next_p = next_p
```

```python
class singleLinkedList: 
    def __init__(self): 
        self.head = None #moramo
        self.tail = None #opciono
        self.size = 0 

    def __len__(self): 
        return self.size 
    def __iter__(self): 
        current = self.head 
        while current is not None: 
            yield current.data 
            current = current.next_p 

    def _iter_node(self): 
        current = self.head 
        while current is not None: 
            yield current
            current = current.next_p 

    def show(self): 
        result = "[" 
        counter = 0 
        for element in self: 
            result += str(element) 
            if counter < len(self) - 1: 
                result += "->" 
            counter += 1 
        return result + "]" 

    def insertEnd(self, data): 

        new_node = Node(data) 

        if self.size == 0: 
            self.head = new_node 
        else: 
            self.tail.next_p = new_node 
        self.tail = new_node 
        self.size += 1  

    def insertStart(self, data): 
        new_node = Node(data, self.head) 

        if self.size == 0: 
            self.tail = new_node 
        self.head = new_node 
        self.size += 1 
        
    def removeStart(self): 
        if self.size == 0: 
            return 
        elif self.size == 1: 
            self.head = None
            self.tail = None 
        self.head = self.head.next_p 
        self.size -= 1 

    def removeEnd(self): 

        if self.size == 0: 
            return 
        elif self.size == 1: 
            self.head = None 
            self.tail = None 
        else: 
            current = self.head 
            while current.next_p != self.tail: 
                current = current.next_p 
            current.next_p = None 
            self.tail = current 
        self.size -= 1

    def removeByValue(self, value): 
        
        if self.size == 0: 
            raise IndexError("List is empty") 
        if self.size == 1 and self.head.data == value: 
            self.head = self.tail = None 
            self.size -= 1 
            return 
        counter = 0 
        prev = None 
        for node in self._iter_node(): 
            if node.data == value and counter == 0:
                self.head = self.head.next_p 
                self.size -= 1 
                return
            if node.data == value and counter == len(self) - 1: 
                self.tail = prev 
                prev.next_p = None 
                self.size -= 1 
                return 
            if node.data == value: 
                prev.next_p = node.next_p 
                self.size -= 1 
                return 
            prev = node 
            counter += 1
        raise ValueError("There is no such element is list")

    def removeByIndex(self, i = None): 

        if i is None: 
            i = self.size - 1 
        if not 0 <= i and i < self.size: 
            raise IndexError("out of bounds") 
        if self.size == 0: 
            raise IndexError("List is empty") 

        counter = 0 
        prev = None 

        for node in self._iter_node(): 
            if i == counter == 0: 
                self.head = self.head.next_p 
                self.size -= 1 
                return  
            if i == counter == len(self) - 1:
                self.tail = prev 
                prev.next_p = None 
                self.size -= 1 
                return   
            if i == counter: 
                prev.next_p = node.next_p 
                self.size -= 1 
                return 
            prev = node 
            counter += 1  

    def first(self): 
        if self.size == 0: 
            raise IndexError("List is empty") 
        return self.head.data
```

```python
class Car: 
    def __init__(self, brand, model, max_speed): 
        self.brand = brand 
        self.model = model 
        self.max_speed = max_speed  

    def __eq__(self, other): 
        if isinstance(other, Car): 
            if self.max_speed == other.max_speed: 
                return True 
            return False 
    # Operato >=
    def __ge__(self, other): 
        if isinstance(other, Car): 
            if self.max_speed >= other.max_speed: 
                return True 
            return False 
```

```python
new_list = singleLinkedList() 
new_list.insertEnd(Car("Tesla", "Model 3", 250)) 
new_list.insertEnd(Car("Tesla", "Model Y", 200)) 
new_list.insertEnd(Car("Tesla", "Model X", 350)) 
```

```python
def find_fastest(): 

    fastest = new_list.first() 

    for car in new_list: 
        if car >= fastest: 
            fastest = car 
    return fastest
```
