## Insert Sort

Pseudo kod

Izvrsni kod

```python
def insert_sort(nasa_lista): 
    n = len(nasa_lista) 

    for i in range(1, n): 

        kljuc = nasa_lista[i] 
        j = i - 1 

        while j >= 0 and nasa_lista[j] > kljuc: 
            nasa_lista[j + 1] = nasa_lista[j] 
            j -= 1 
        nasa_lista[j + 1] = kljuc
```

```python
nasa_lista = [55,22,44,33,11,66] 
print("Originalna lista", nasa_lista) 
insert_sort(nasa_lista) 
print("Sortirana lista", nasa_lista)
```
